```python
operation0 = pd.read_csv('operations.csv', index_col='date_operation', parse_dates=True)
operation = operation0.copy()
operation
```

```python
operation.columns
```

```python
operation = operation.drop(['libelle', 'categ'], axis=1)
```

```python
operation = operation.dropna(axis=0)
```

```python
operation.loc[ -operation['montant'] > operation['solde_avt_ope'] , 'montant'] = 0
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
operation.loc[:,'solde_avt_ope'].resample('M').plot()
plt.subplot(2,1,2)
operation.loc[:,'montant'].resample('M').plot()

plt.legend(loc='best')
plt.show()
```
